```python
# utilizamos nuestras credenciales 
token_client_id = os.getenv('SPOTIPY_CLIENT_ID')
token_client_secret = os.getenv('SPOTIPY_CLIENT_SECRET')
```

```python
sp = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials
                     (client_id=token_client_id,
                      client_secret=token_client_secret))
```

```python
# probamos la api con un artista para conseguir el género y nos devuelve un json()
def get_genres(track_id):
    track_info = sp.track(track_id)
    return sp.artist(track_info["artists"][0]["id"])["genres"]
```

```python
# generamos una función para conseguir los followers 
def get_followers(track_id):
    track_info = sp.track(track_id)
    return sp.artist(track_info["artists"][0]["id"])["followers"]["total"]
```
