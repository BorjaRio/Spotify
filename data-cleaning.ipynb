### Abrimos y exploramos nuestros dataset (data)

```python
data = pd.read_csv("dataset/data_enrinch.csv")
```

```python
data.head(5)
```

```python
# miramos el tipo de columnas y el tipo de datos que tenemos
data.info()
```

```python
# comprobamos los valores nulos
data.isnull().sum()
```

```python
data.columns
```

```python
data_r=data.round(decimals=3)
data_r.head(5)
```

```python
# quitamos las columnas que no vamos a necesitar, quitamos la columna id ya que ahora identificaremos a los artistas por la columna artits
data_r=data_r.drop(columns=['id'])
```

```python
data_r.tail(5)
```

```python
data_r["artists"]= data_r["artists"].apply(lambda x: x.replace("[","").replace("]","").replace("'","").split(","))
```

```python
data_r.head(5)
```

```python
data_explode=data_r.explode("artists")
data_explode.info()
```

```python
data_explode_g=data_explode.explode("genre")
data_explode_g.head(5)
```

```python
# quitamos los corchetes de genero
data_explode_g["genre"]= data_explode_g["genre"].apply(lambda x: x.replace("[","").replace("]","").replace("'","").split(","))
```

```python
data_d=data_explode_g.explode("genre")
```

```python
data_d.head(5)
```

```python
# vemos que de las 227815 datos, 23k tienen valores vacios por lo que los sustituimos por na
data_d["genre"].value_counts().head(20)
```

```python
# exploramos la columna followers, vemos que tenemos un nulo
data_explode_g["followers"].isnull().sum()
```

```python
## rellenamos el valor nullo con un cero
data_cleaned=data_explode_g.fillna(value=0, axis=1)
data_cleaned.isnull().sum()
```

```python
data_cleaned["followers"].max()
```

```python
data_cleaned["followers"].min()
```

```python
### Exportamos nuestro csv una vez estÃ¡ limpio para la visualizaciÃ³n de datos
data_d.to_csv("data_cleaned.csv")
```

### Artistas mas populares por suma

```python
plt.figure(figsize=(15, 10))
sns.set(style="whitegrid")
x = data.groupby("artists")["popularity"].sum().sort_values(ascending=False).head(20)
ax = sns.barplot(x=x, y=x.index, data=data, palette="Blues_d")
ax.set_title('Top Artists mÃ¡s populares por suma')
ax.set_ylabel('Popularidad')
ax.set_xlabel('Artista')
plt.xticks(rotation = 90)
```
